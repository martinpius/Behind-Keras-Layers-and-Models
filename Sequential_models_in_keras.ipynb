```
#We can also stack layers ontop of each other like as follow
#input = tf.keras.layers.Input(shape = [10,100], name = 'input_shape')
layer1 = tf.keras.layers.Dense(units = 64, kernel_initializer= 'random_normal',name = 'layer_1')
layer2 = tf.keras.layers.Activation(tf.nn.relu)
layer3 = tf.keras.layers.Dense(units = 32, kernel_initializer= 'random_normal', name = 'layer_3')
layer4 = tf.keras.layers.Activation(tf.nn.relu)
layer5 = tf.keras.layers.Dense(units = 4, kernel_initializer = 'random_normal', name = 'output')
layer6 = tf.keras.layers.Activation(tf.nn.softmax)
model = layer6(layer5(layer4(layer3(layer2(layer1(tf.ones(shape = (3,3))))))))
```

```
from tensorflow.keras.layers import Dense, Activation,MaxPooling2D, GlobalMaxPooling2D, BatchNormalization, Conv2D, Input, Dropout
```

```
from tensorflow.keras.models import Sequential
```

```
model = Sequential()
```

```
model.add(Input(shape = (250,250,3), name = 'input'))
```

```
model.add(Conv2D(filters = 32, kernel_size = (5,5), strides = (2,2), padding = 'valid'))
```

```
model.add(Activation('relu'))
```

```
model.add(Conv2D(filters = 32, kernel_size =(5,5), strides = (2,2), padding = 'valid'))
model.add(MaxPooling2D(pool_size = 3))
```

```
model.add(BatchNormalization())
```

```
model.summary()
```

```
model.add(Conv2D(filters =32, kernel_size = (3,3), strides = (2,2),padding = 'same'))
```

```
model.add(Activation('relu'))
```

```
model.add(MaxPooling2D(pool_size = 3))
```

```
model.add(BatchNormalization())
```

```
model.summary()
```

```
model.add(Conv2D(filters = 32, kernel_size =(2,2), strides = (1,1), padding = 'same'))
```

```
model.add(Activation('relu'))
```

```
model.add(BatchNormalization())
```

```
model.add(MaxPooling2D(pool_size = (3)))
```

```
model.add(GlobalMaxPooling2D())
```

```
model.add(Dense(10, activation = 'relu'))
```

```
model.summary()
```

```
display([layer.output for layer in model.layers])#Displaye all layers in the model.
```

```
#Sample Model to reuse
model = Sequential([Input(shape = (250,250,3)),
Conv2D(filters = 32, kernel_size = (5,5),strides = (2,2),padding = 'same',activation = 'relu'),
Conv2D(filters = 32, kernel_size = (3,3), strides = (2,2), padding = 'same', activation = 'relu'),
BatchNormalization(),
Dropout(rate = 0.2),
MaxPooling2D(pool_size = 3),
Conv2D(filters = 32, kernel_size = (3,3), strides = (1,1),padding = 'valid', activation = 'relu'),
])
#Extracts features from the above model and reuse it in another keras model
fetures = tf.keras.Model(inputs = model.inputs, outputs = [layer.output for layer in model.layers])
```

```
display(fetures)
```

```
input_test = tf.ones((100, 250,250,3))
```

```
#Call our feature's extract
fetures(input_test)
```

```
#Feature extraction from a specified layer
model = Sequential([
                    Input(shape = (128,128,3), name = 'input_shape'),
                    Conv2D(filters = 32, kernel_size = (5,5), strides = (2,2), padding = 'same', activation = 'relu', name = 'first_layer'),
                    Conv2D(filters = 32, kernel_size = (3,3), strides = (1,1), padding = 'same', activation ='relu', name = 'second_layer'),
                    BatchNormalization(),
                    MaxPooling2D(pool_size = 3),
                    Dropout(rate = 0.2),
          
])
```

```
feat_spec_extract = tf.keras.Model(
    inputs = model.inputs, outputs = model.get_layer(name = 'second_layer').output
)
```

```
#Call the extract_spec above
feat_spec_extract(tf.ones(shape = (2,250,250,3)))
```

```
##The above procedure of feature's extraction is a good step towards transfer learning in keras sequential models
#A simple architecture of a transfer learning
#Load model
mymodel = tf.keras.applications.InceptionV3(weights= 'imagenet',pooling = 'max', include_top = False)
```
